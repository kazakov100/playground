```python
import os
import json
from dotenv import load_dotenv
import requests

# Load environment variables
load_dotenv()
token = os.getenv("GBFS_TOKEN")

url = "https://mds.bird.co/gbfs/v2/public/barcelona/station_information.json"


headers = {
    "Authorization": f"Bearer {token}",
    "Accept": "application/json"
}

r = requests.get(url, headers=headers)
print(r.status_code)

# Pretty JSON output
try:
    data = r.json()
    print(json.dumps(data, indent=2))
except Exception as e:
    print("Error parsing JSON:", e)
    print(r.text)
```
